```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline

df = pd.read_csv("../data/colleges.csv")
df.head()
```

```python
df.shape
df.info()
```

```python
df = df.rename(columns={
    "INSTNM": "college_name",
    "CITY": "city",
    "STABBR": "state",
    "CONTROL": "college_type",
    "UGDS": "undergrad_enrollment",
    "TUITIONFEE_IN": "tuition_in_state",
    "TUITIONFEE_OUT": "tuition_out_state",
    "ADM_RATE": "admission_rate",
    "C150_4": "graduation_rate"
})
```

```python
numeric_cols = [
    "undergrad_enrollment",
    "tuition_in_state",
    "tuition_out_state",
    "admission_rate",
    "graduation_rate"
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df["college_type"] = df["college_type"].map({
    1: "Public",
    2: "Private Non-Profit",
    3: "Private For-Profit"
})
```

```python
sns.countplot(y="college_type", data=df)
plt.title("Distribution of Colleges by Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="college_type", y="tuition_in_state", data=df)
plt.title("In-State Tuition by College Type")
plt.show()
```

```python
state_counts = df["state"].value_counts().head(10)

sns.barplot(x=state_counts.values, y=state_counts.index)
plt.title("Top 10 States by Number of Colleges")
plt.xlabel("Number of Colleges")
plt.show()
```

```python
sns.histplot(df["undergrad_enrollment"], bins=30, kde=True)
plt.title("Undergraduate Enrollment Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="tuition_in_state",
    y="graduation_rate",
    hue="college_type",
    data=df,
    alpha=0.7
)
plt.title("Tuition vs Graduation Rate")
plt.show()
```

```python
fig = px.scatter(
    df,
    x="tuition_in_state",
    y="graduation_rate",
    color="college_type",
    size="undergrad_enrollment",
    hover_name="college_name",
    title="Interactive Tuition vs Graduation Rate Dashboard"
)

fig.show()
```

```python
fig_map = px.scatter_geo(
    df,
    locations="state",
    locationmode="USA-states",
    color="college_type",
    hover_name="college_name",
    scope="usa",
    title="Geographic Distribution of U.S. Colleges"
)

fig_map.show()
```

```python
## Key Insights

- Public colleges generally have significantly lower in-state tuition compared to private institutions.
- States like California, Texas, and New York have the highest concentration of colleges.
- Higher tuition does not always guarantee higher graduation rates.
- Colleges with larger enrollments tend to show wider variability in graduation outcomes.
```
